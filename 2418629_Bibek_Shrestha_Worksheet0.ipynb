###Task 1: Create a Python program that converts between different units of measurement.

```python
def convert_units():
    """
    A program that converts between different units of measurement.
    Supports length (m/ft), weight (kg/lbs), and volume (L/gal).
    """
    print("Choose a conversion type:")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kilograms <-> pounds)")
    print("3. Volume (liters <-> gallons)")

    while True:
        conversion_type = input("Enter your choice (1/2/3): ")
        if conversion_type in ['1', '2', '3']:
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

    while True:
        try:
            value = float(input("Enter the value to convert: "))
            break
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

    result = None
    unit_from = ''
    unit_to = ''

    if conversion_type == '1':  # Length conversion
        print("\nChoose length conversion:")
        print("a. Meters to Feet")
        print("b. Feet to Meters")
        while True:
            length_choice = input("Enter your choice (a/b): ").lower()
            if length_choice in ['a', 'b']:
                break
            else:
                print("Invalid choice. Please enter 'a' or 'b'.")
        if length_choice == 'a':
            result = value * 3.28084
            unit_from = 'meters'
            unit_to = 'feet'
        else:
            result = value / 3.28084
            unit_from = 'feet'
            unit_to = 'meters'
    elif conversion_type == '2':  # Weight conversion
        print("\nChoose weight conversion:")
        print("a. Kilograms to Pounds")
        print("b. Pounds to Kilograms")
        while True:
            weight_choice = input("Enter your choice (a/b): ").lower()
            if weight_choice in ['a', 'b']:
                break
            else:
                print("Invalid choice. Please enter 'a' or 'b'.")
        if weight_choice == 'a':
            result = value * 2.20462
            unit_from = 'kilograms'
            unit_to = 'pounds'
        else:
            result = value / 2.20462
            unit_from = 'pounds'
            unit_to = 'kilograms'
    elif conversion_type == '3':  # Volume conversion
        print("\nChoose volume conversion:")
        print("a. Liters to Gallons")
        print("b. Gallons to Liters")
        while True:
            volume_choice = input("Enter your choice (a/b): ").lower()
            if volume_choice in ['a', 'b']:
                break
            else:
                print("Invalid choice. Please enter 'a' or 'b'.")
        if volume_choice == 'a':
            result = value * 0.264172
            unit_from = 'liters'
            unit_to = 'gallons'
        else:
            result = value / 0.264172
            unit_from = 'gallons'
            unit_to = 'liters'

    if result is not None:
        print(f"\n{value:.2f} {unit_from} is equal to {result:.2f} {unit_to}.")
    else:
        print("An unexpected error occurred during conversion.")


convert_units()
```

###Task 2: Create a Python program that performs various mathematical operations on a list of numbers.

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The sum of the numbers.
    """
    return sum(numbers)

def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The average of the numbers.

    Raises:
        ValueError: If the input list is empty.
    """
    if not numbers:
        raise ValueError("Cannot calculate average of an empty list.")
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The maximum value in the list.

    Raises:
        ValueError: If the input list is empty.
    """
    if not numbers:
        raise ValueError("Cannot find maximum of an empty list.")
    return max(numbers)

def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The minimum value in the list.

    Raises:
        ValueError: If the input list is empty.
    """
    if not numbers:
        raise ValueError("Cannot find minimum of an empty list.")
    return min(numbers)

def mathematical_operations_program():
    """
    A program that performs various mathematical operations on a list of numbers.
    Supports sum, average, maximum, and minimum.
    """
    print("Choose an operation:")
    print("1. Find Sum")
    print("2. Find Average")
    print("3. Find Maximum")
    print("4. Find Minimum")

    while True:
        choice = input("Enter your choice (1/2/3/4): ")
        if choice in ['1', '2', '3', '4']:
            break
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 4.")

    while True:
        numbers_input = input("Enter a list of numbers separated by spaces: ")
        try:
            numbers = [float(x) for x in numbers_input.split()]
            if not numbers and choice != '0':
                print("The list of numbers cannot be empty for these operations. Please enter some numbers.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter numbers separated by spaces.")

    result = None
    operation_name = ""

    try:
        if choice == '1':
            result = calculate_sum(numbers)
            operation_name = "Sum"
        elif choice == '2':
            result = calculate_average(numbers)
            operation_name = "Average"
        elif choice == '3':
            result = find_maximum(numbers)
            operation_name = "Maximum"
        elif choice == '4':
            result = find_minimum(numbers)
            operation_name = "Minimum"

        if result is not None:
            print(f"The {operation_name} of {numbers_input} is: {result:.2f}")

    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

mathematical_operations_program()
```

###Task 1:Write a Python function that extracts every other element from a list, starting from the first element.

```python
def extract_every_other(lst):
    """
    Extracts every other element from a list, starting from the first element.

    Args:
        lst (list): The input list.

    Returns:
        list: A new list containing every other element.
    """
    return lst[::2]

first_list = [1, 2, 3, 4, 5, 6]
extracted_elements = extract_every_other(first_list)
print(f"Original list: {first_list}")
print(f"Every other element (starting from the first): {extracted_elements}")

second_list = ['a', 'b', 'c', 'd', 'e']
extracted_elements_2 = extract_every_other(second_list)
print(f"\nOriginal list: {second_list}")
print(f"Every other element (starting from the first): {extracted_elements_2}")
```

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from a given list, starting from a specified index
    and ending at another specified index (inclusive).

    Args:
        lst (list): The input list.
        start (int): The starting index (inclusive).
        end (int): The ending index (inclusive).

    Returns:
        list: The sublist from 'start' to 'end' (inclusive).
    """
    return lst[start : end + 1]

first_list = [1, 2, 3, 4, 5, 6]
start_index = 2
end_index = 4
sublist = get_sublist(first_list, start_index, end_index)

print(f"Original list: {first_list}")
print(f"Sublist from index {start_index} to {end_index}: {sublist}")

second_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
start_index_2 = 1
end_index_2 = 5
sublist_2 = get_sublist(second_list, start_index_2, end_index_2)

print(f"\nOriginal list: {second_list}")
print(f"Sublist from index {start_index_2} to {end_index_2}: {sublist_2}")
```

##Exercise on Nested List

###Task 1: Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Args:
        lst (list): A nested list.

    Returns:
        list: A flattened list.
    """
    flat_list = []

    for item in lst:
        for element in item:
            flat_list.append(element)

    return flat_list


nested_list = [[1, 2], [3, 4], [5]]
flattened = flatten(nested_list)

print(f"Original list: {nested_list}")
print(f"Flattened list: {flattened}")
```

###Task 2: Write a Python function that extracts a specific element from a nested list given its indices.

```python
def access_nested_element(lst, indices):
    """
    Accesses an element from a nested list using indices.

    Args:
        lst (list): A nested list.
        indices (list): A list of indices.

    Returns:
        element: The element at the specified position.
    """
    element = lst

    for index in indices:
        element = element[index]

    return element


lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
result = access_nested_element(lst, [1, 2])

print(f"\nNested list: {lst}")
print(f"Accessed element: {result}")
```

###Task 3: Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

```python
def sum_nested(lst):
    """
    Calculates the sum of all elements in a nested list.

    Args:
        lst (list): A nested list.

    Returns:
        int: Sum of all numbers.
    """
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item

    return total


nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)

print(f"\nNested list: {nested_list}")
print(f"Sum of elements: {result}")
```

###Task 4: Write a Python function that removes all occurrences of a specific element from a nested list.

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of an element from a nested list.

    Args:
        lst (list): A nested list.
        elem (any): Element to remove.

    Returns:
        list: Modified list without the element.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)

    return result


lst = [[1, 2], [3, 2], [4, 5]]
result = remove_element(lst, 2)

print(f"\nOriginal list: {lst}")
print(f"List after removing 2: {result}")
```

###Task 5: Write a Python function that finds the maximum element in a nested list (regardless of depth).

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Args:
        lst (list): A nested list.

    Returns:
        int: Maximum element.
    """
    max_value = float('-inf')

    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))
        else:
            max_value = max(max_value, item)

    return max_value


nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)

print(f"\nNested list: {nested_list}")
print(f"Maximum element: {result}")
```

###Task 6: Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurrences(lst, elem):
    """
    Counts occurrences of an element in a nested list.

    Args:
        lst (list): A nested list.
        elem (any): Element to count.

    Returns:
        int: Number of occurrences.
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1

    return count


lst = [[1, 2], [2, 3], [2, 4]]
result = count_occurrences(lst, 2)

print(f"\nNested list: {lst}")
print(f"Occurrences of 2: {result}")
```

###Task 7: Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list regardless of depth.

    Args:
        lst (list): A deeply nested list.

    Returns:
        list: A completely flattened list.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result


nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)

print(f"\nOriginal list: {nested_list}")
print(f"Flattened list: {result}")
```

###Task 8: Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):
    """
    Calculates the average of elements in a nested list.

    Args:
        lst (list): A nested list.

    Returns:
        float: Average of all numbers.
    """
    total = 0
    count = 0

    def helper(sublist):
        nonlocal total, count
        for item in sublist:
            if isinstance(item, list):
                helper(item)
            else:
                total += item
                count += 1

    helper(lst)

    return total / count


nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)

print(f"\nNested list: {nested_list}")
print(f"Average value: {result}")
```

```python
# Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Mea-
# sure and print the time taken for this operation.
size = 1000

A = [[1 for _ in range(size)] for _ in range(size)]
B = [[1 for _ in range(size)] for _ in range(size)]

start = time.time()

result = [[0]*size for _ in range(size)]

for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()

print("Python List Matrix Multiplication Time:", end - start, "seconds")

# Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
# Measure and print the time taken for this operation.
A = np.ones((1000,1000))
B = np.ones((1000,1000))

start = time.time()

result = np.dot(A,B)

end = time.time()

print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
